해싱에서 충돌(Collision)이 발생하면 일반적으로 개방 주소법(Open Addressing)과 체이닝(Chaining) 두 가지 방법으로 해결한다.

개방 주소법은 충돌이 발생했을 때 해시 테이블 내부의 다른 빈 공간을 찾아 저장하는 방식이다. 대표적으로 선형 탐색, 이차원 탐색, 더블 해싱 등이 있다.

장점은 추가적인 연결 리스트가 필요 없어서 메모리 사용량이 적고, 배열 내부에 데이터가 저장되므로 캐시 효율이 좋다는 점이다. 또한 구현이 비교적 단순하다.

단점은 충돌이 많아질수록 탐색 시간이 증가하고, 군집화(Clustering) 문제가 발생할 수 있다는 점이다. 또한 테이블이 거의 가득 차면 성능이 급격히 저하되며 삭제 구현도 복잡하다.

체이닝은 하나의 해시 주소에 여러 데이터를 연결 리스트 형태로 저장하는 방식이다.

장점은 충돌이 발생해도 저장 공간만 충분하면 계속 저장할 수 있으며 삭제가 쉽다는 점이다. 또한 충돌이 증가해도 성능 저하가 상대적으로 적다.

단점은 연결 리스트를 위한 추가 메모리가 필요하고, 포인터 사용으로 인해 접근 속도가 느려질 수 있다는 점이다. 또한 연결 리스트 길이가 길어질수록 검색 시간이 증가한다.

정리하면 개방 주소법은 메모리 효율이 좋지만 충돌 증가 시 성능 저하가 크고, 체이닝은 메모리를 더 사용하지만 충돌에 더 안정적인 방법이다.

하지만 검색에서는 효율이 좋아질 수 있다.

실패하는 검색의 경우에도 현재 값이 찾는 값보다 커지는 순간 탐색을 중단할 수 있으므로 성능이 향상된다.

삭제 또한 원하는 값을 더 빨리 찾을 수 있어서 약간 효율이 좋아질 수 있다.

즉, 정렬된 체이닝은 삽입 성능은 감소하지만 검색 성능은 향상되는 특징이 있다.
